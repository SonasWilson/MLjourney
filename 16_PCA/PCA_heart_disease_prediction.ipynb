Treating outliers


---


. Usual guideline is to remove anything that has Z score > 3 formula or Z score < -3

```python
df[df.Cholesterol>(df.Cholesterol.mean()+3*df.Cholesterol.std())]
```

```python
df.shape
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df[df.MaxHR>(df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df[df.FastingBS>(df.FastingBS.mean()+3*df.FastingBS.std())]
```

```python
df[df.Oldpeak>(df.Oldpeak.mean()+3*df.Oldpeak.std())]
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df.ChestPainType.unique()
```

```python
df.RestingECG.unique()
```

```python
df.ExerciseAngina.unique()
```

```python
df.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True).astype(int)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, y_train)
model_dt.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
model_svm = SVC(kernel='linear', gamma=100, C=1)
model_svm.fit(X_train, y_train)
model_svm.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
model_lr.score(X_test, y_test)
```

Best score given by Random Forest

Using PCA to reduce dimensions

```python
X.shape
```

```python
X.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```

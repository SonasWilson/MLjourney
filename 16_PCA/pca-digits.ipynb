. PCA finds the best features/components that have high impact on target

. PCA is also called dimensionality reduction technique which help to deal with the dimensionality curse

Few things to keep in mind:


*   accuracy may drop
*   scale features before applying PCA

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = digits.target
```

**Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

Creating and training model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

Using PCA to reduce dimensions

```python
X.shape
```

Use components such that 95% of variance is retained

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

***PCA created 29 components out of 64 original columns***

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

accuraccy decreased with slight difference

**Selecting only two components**

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

**both combined retains 0.14+0.13=0.27 or 27% of important feature information**

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

Got less accuancy (~60%) as using only 2 components did not retain much of the feature information.

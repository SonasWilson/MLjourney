```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

### Reading and plotting the data

```python
df = pd.read_csv('canada_per_capita_income.csv', delimiter=',')
df.head()
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df['per capita income (US$)'], marker="+", color="green")
plt.show()
```

### Setting up the model

```python
year = df.drop('per capita income (US$)', axis='columns')
year
```

```python
income = df['per capita income (US$)']
income.head()
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(year, income)
```

```python
m = lin_reg.coef_
m
```

```python
c = lin_reg.intercept_
c
```

### Predict for the year 2020

```python
lin_reg.predict([[2020]])
```

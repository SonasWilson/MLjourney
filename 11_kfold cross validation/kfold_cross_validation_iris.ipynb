```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
from sklearn.model_selection import cross_val_score
```

**1. Logistic Regression**

```python
l_score = cross_val_score(LogisticRegression(random_state=42, max_iter=200), iris.data, iris.target)
l_score
```

```python
from numpy import average
average(l_score)
```

**2. SVM**

```python
svm_score = cross_val_score(SVC(gamma=1, kernel='linear', C=1), iris.data, iris.target)
svm_score
```

```python
average(svm_score)
```

```python
svm_score = cross_val_score(SVC(gamma=1, kernel='poly', C=1), iris.data, iris.target)
svm_score
```

```python
average(svm_score)
```

**3. Random Forest**

```python
rf_score = cross_val_score(RandomForestClassifier(n_estimators=40,), iris.data, iris.target)
rf_score
```

```python
average(rf_score)
```

**Decision tree**

```python
d_score = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target)
d_score
```

```python
average(d_score)
```

**Best score so far is from SVM: 0.9800000000000001**


*   SVM > Logistic Regression > Decision Tree ~ Random Forest

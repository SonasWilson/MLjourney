**label encoding and one hot encoding**

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=20)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Train a model using standalone SVM**

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), X, y, cv=5)
scores.mean()
```

**Using bagging**

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(estimator=SVC(), n_estimators=100, max_samples=0.8, random_state=0)
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

not much improvement

**Train a model using standalone decision tree**

```python
from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(random_state=0), X, y, cv=5)
scores.mean()
```

**Using bagging**

```python
bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=10),
    n_estimators=100,
    max_samples=0.9,
    oob_score=True,
    random_state=0
)

scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

**With bagging the score improved from 71.93% to 80.37%**

# **Train a model using Random Forest which itself uses bagging**

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), X, y, cv=5)
scores.mean()
```

**Random forest gave even a better performance with 81.7% as score**
